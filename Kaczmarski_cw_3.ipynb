1. Analiza grupowania ma na celu oddzielenie grupy osób chorych i zdrowych, oraz określenie czy wyniki będą w stanie na ten podział wskazać.
2. Zmiennymi objaśniającymi są kolumny od 2 do 32, czyli radius_mean, texture_mean itd. Nazwy (numery) skupień zawiera kolumna 2 (indeks 1) 'diagnosis' o wartościach M lub B. Będzie ona wykorzystana do porównania trafności grupowania algorytmu. W obliczeniach pomijamy pierwszą kolumnę bowiem jest to numer ID pacjenta nie dostarczający nam danych do grupowania.

```
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data')

y=df.values[:,1]
le=LabelEncoder()
y=le.fit_transform(y)
X=df.values[:,2:32]
print(X)

y_kmeans = KMeans(n_clusters=2).fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
plt.show()


print('y_true: ')
print(y)
print('y_kmeans: ')
print(y_kmeans)
print('Accuracy k-means: ', accuracy_score(y,y_kmeans))
```

5. Metoda K-średnich skutecznie oddzieliła zgrupowania. Accuracy wyniosło ok. 85%. Na wykresie widać rozdzielone skupienia lekko się nachodzące. Wynik zastosowania metody jest zgodny z oczekiwaniami.
